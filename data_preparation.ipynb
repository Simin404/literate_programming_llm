```python
import pandas as pd

df = pd.read_parquet("hf://datasets/christopher/rosetta-code/data/train-00000-of-00001-8b4da49264116bbf.parquet")


# import pandas as pd

# df = pd.read_parquet("hf://datasets/pharaouk/rosetta-code/data/train-00000-of-00001-8b4da49264116bbf.parquet")
```

```python
train_df['normalized_task'] = train_df['task'].str.replace(r'[^a-zA-Z0-9]', '', regex=True).str.lower()
test_df['normalized_task'] = test_df['task'].str.replace(r'[^a-zA-Z0-9]', '', regex=True).str.lower()

df['normalized_task_name'] = df['task_name'].str.replace(r'[^a-zA-Z0-9]', '', regex=True).str.lower()
```

```python
df = df.drop_duplicates(subset=['normalized_task_name'])
df.shape
```

```python
train_df
```

```python
df[['normalized_task_name', 'task_description']]
```

```python
merged_df = pd.merge(
    train_df,
    df[['normalized_task_name', 'task_description']],  # Select only the necessary columns
    left_on='normalized_task',
    right_on='normalized_task_name',
    how='left'
)
```

```python
merged_df.shape
```

```python
merged_df_2 = pd.merge(
    test_df,
    df[['normalized_task_name', 'task_description']],  # Select only the necessary columns
    left_on='normalized_task',
    right_on='normalized_task_name',
    how='left'
)
```

```python
merged_df_2.shape
```

```python
## Add missing description manually

# text1 = pd.read_csv('text1.txt', header=None, names=['Name'])
with open('text1.txt', 'r') as file:
    content = file.read()
merged_df.loc[merged_df['task'] == 'Continued-fraction-Arithmetic-G-matrix-ng-continued-fraction-n-', 'task_description'] = content
merged_df_2.loc[merged_df_2['task'] == 'Continued-fraction-Arithmetic-G-matrix-ng-continued-fraction-n-', 'task_description'] = content


# text2 = pd.read_csv('text2.txt', header=None, names=['Name'])
with open('text2.txt', 'r') as file:
    content = file.read()
merged_df.loc[merged_df['task'] == 'Events', 'task_description'] = content
merged_df_2.loc[merged_df_2['task'] == 'Events', 'task_description'] = content


# text3 = pd.read_csv('text3.txt', header=None, names=['Name'])
with open('text3.txt', 'r') as file:
    content = file.read()
merged_df.loc[merged_df['task'] == 'Gotchas', 'task_description'] = content
merged_df_2.loc[merged_df_2['task'] == 'Gotchas', 'task_description'] = content


# text4 = pd.read_csv('text4.txt', header=None, names=['Name'])
with open('text4.txt', 'r') as file:
    content = file.read()
merged_df.loc[merged_df['task'] == 'Mutex', 'task_description'] = content
merged_df_2.loc[merged_df_2['task'] == 'Mutex', 'task_description'] = content
```

```python
train_df['task_description'] = merged_df['task_description']
test_df['task_description'] = merged_df_2['task_description']
```
